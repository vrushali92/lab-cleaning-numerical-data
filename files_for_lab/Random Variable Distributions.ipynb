# Check if any of the transformations (log-transform, etc.) we have seen up to this point changes the result.

```python
def log_transfom_clean(x):
    if x>0:
        return np.log(x)
    else:
        return np.NAN
```

```python
numerical_df.columns
```

```python
sns.displot(numerical_df.income.apply(log_transfom_clean), kde = True)
```

```python
sns.displot(numerical_df.monthly_premium_auto.apply(log_transfom_clean), kde = True)
```

```python
sns.displot(numerical_df.months_since_last_claim.apply(log_transfom_clean), kde = True)
```

```python
sns.displot(numerical_df.months_since_policy_inception.apply(log_transfom_clean), kde = True)
```

```python
sns.displot(numerical_df.number_of_open_complaints.apply(log_transfom_clean), kde = True)
```

```python
sns.displot(numerical_df.total_claim_amount_transformed, kde = True)
```

```python
sns.displot(numerical_df.customer_lifetime_value_transformed, kde = True)
```
